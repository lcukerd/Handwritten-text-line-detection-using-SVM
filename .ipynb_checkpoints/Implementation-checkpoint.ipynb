```python
def connectSeparatorsH(pRegions, delta, SPR, h):
    textRegions = [];
    associateN = np.ones((h)) * -1;
    once = False;
    for i in range(len (pRegions)):
        regionsStrip = pRegions[i];
        index = 0;

        if delta[i] == 0:
            textRegions.append([-1]);
            continue;
        if i + 1 < len (pRegions):
            regionsStripN = pRegions[i + 1];
        else:
            regionsStripN = -1;

        associateT = np.copy(associateN);
        associateN = np.ones((h)) * -1;
        textLines = [];
        for j in range(len(regionsStrip)):
            region = regionsStrip[j];
            if associateT[j] == -1 and once == True:
                for k in range(i - 1, -1, -1):
                    l = findPreviousSeparator(regionsStrip, j - 1, pRegions[k]) if j - 1 >= 0 else -1;
                    r = findPreviousSeparator(regionsStrip, j + 1, pRegions[k]) if j + 1 < len(regionsStrip) else len(pRegions[k]) + 1;
                    
                    y = minimize(region, SPR[k], pRegions[k], l, r);
                    success = False
                    
                    for pos in range(len(pRegions[k])):
                        regionI = pRegions[k][pos];
                        if regionI == y:
                            print ("Already Present", k, region, regionI)
                            success = True;
                            break;
                        elif regionI > y:
                            success = True;
                            backup = pRegions[k][pos:];
                            pRegions[k][pos] = y;
                            pRegions[k][pos + 1:] = backup;
                            break;
                    
                    if success == False:
                        pRegions[k].append(y);
                
            if regionsStripN == -1:
                continue;
                
            index = findNextSeparator(regionsStrip[j], regionsStripN);
            
            if associateN[index] == 1:
                regionsStripN = performMultiAssociation(regionsStrip, j, regionsStripN, index, SPR[i + 1]);
                associateN[index + 1] = 1;
                index += 1;
            else:
                associateN[index] = 1;
            textLines.append(index);
        textRegions.append(textLines);
        once = True
    return textRegions;

def findNextSeparator(region, regionsStripN):
    pos = 0;
    while pos + 1 < len(regionsStripN) and abs(region - regionsStripN[pos]) > abs(region - regionsStripN[pos + 1]):
        pos += 1;
    
    return pos;

def findPreviousSeparator(regionsStrip, index, regionsStripP):
    for i in range(len(regionsStripP)):
        if index == findNextSeparator(regionsStripP[i], regionsStrip):
            return i;
    return -1;
    
```
